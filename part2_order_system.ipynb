```python
categories = ["Starters", "Mains" ,"Desserts"]

for category in categories:
   print(f"\n===== {category} =====")
   for item, details in menu.items():
    if details["category"] == category:
      availability = "Available" if details["available"] else "Unavailable"
      print(f"{item:<18} ₹{details['price']:.2f}   [{availability}]")
```

```python
total_items = len(menu)

total_available = sum(1 for item in menu.values() if item["available"])

most_expensive = max(menu, key=lambda x: menu[x]["price"])

print(f"Total items on menu : {total_items}")
print(f"Total available     : {total_available}")
print(f"Most expensive item : {most_expensive} — ₹{menu[most_expensive]['price']:.2f}")

print("\nItems under ₹150:")
for item, details in menu.items():
  if details["price"] < 150:
    print(f" {item} - ₹{details['price']:.2f}")
```

TASK 2



```python
cart = []

def add_item(cart, item_name, quantity):
    # Check 1 - does item exist in menu?
    if item_name not in menu:
        print(f" '{item_name}' does not exist in the menu.")
        return

    # Check 2 - is it available?
    if not menu[item_name]["available"]:
        print(f" '{item_name}' is currently unavailable.")
        return

    # Check 3 - already in cart? update quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f" Updated '{item_name}' quantity to {entry['quantity']}")
            return

    # Not in cart yet - add new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f" Added '{item_name}' x{quantity}")
```

```python
def remove_item(cart, item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"🗑️ Removed '{item_name}' from cart.")
            return
    print(f"'{item_name}' not found in cart.")

def update_quantity(cart, item_name, new_quantity):
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"'{item_name}' quantity updated to {new_quantity}")
            return
    print(f"'{item_name}' not found in cart.")

# Simulate sequence
add_item(cart, "Paneer Tikka", 2)
print("Cart:", cart)

add_item(cart, "Gulab Jamun", 1)
print("Cart:", cart)

add_item(cart, "Paneer Tikka", 1)
print("Cart:", cart)

add_item(cart, "Mystery Burger", 1)
add_item(cart, "Chicken Wings", 1)

remove_item(cart, "Gulab Jamun")
print("Cart:", cart)

# Order Summary
print("\n========== Order Summary ==========")
subtotal = 0
for entry in cart:
    line_total = entry["quantity"] * entry["price"]
    subtotal += line_total
    print(f"{entry['item']:<20} x{entry['quantity']}    ₹{line_total:.2f}")

gst   = round(subtotal * 0.05, 2)
total = round(subtotal + gst, 2)
print("------------------------------------")
print(f"{'Subtotal:':<25} ₹{subtotal:.2f}")
print(f"{'GST (5%):':<25} ₹{gst:.2f}")
print(f"{'Total Payable:':<25} ₹{total:.2f}")
print("====================================")
```

```python
import copy

inventory_backup = copy.deepcopy(inventory)

inventory["Paneer Tikka"]["stock"] = 999
print("Original:", inventory["Paneer Tikka"]["stock"])
print("Backup:", inventory_backup["Paneer Tikka"]["stock"])
inventory["Paneer Tikka"]["stock"] = 10

for entry in cart:
  item = entry["item"]
  qty = entry["quantity"]
  if inventory[item]["stock"] >= qty:
    inventory[item]["stock"] -= qty
  else:
    print(f"Insufficient stock for {item}! Deducting what's available.")
    inventory[item]["stock"] = 0

print("\n--- Reorder Alerts ---")
for item, details in inventory.items():
  if details["stock"] <= details["reorder_level"]:
      print(f"Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")

print("\nInventory Paneer Tikka stock now:", inventory["Paneer Tikka"]["stock"])
print("Backup Paneer Tikka stock:", inventory_backup["Paneer Tikka"]["stock"])

print("\n--- Final Inventory ---")
for item, details in inventory.items():
    print(f"{item}: stock = {details['stock']}")

print("\n--- Backup Inventory ---")
for item, details in inventory_backup.items():
    print(f"{item}: stock = {details['stock']}")
```

```python
# Part 1 - Revenue per day
print("--- Revenue Per Day ---")
for date, orders in sales_log.items():
    daily_total = sum(order["total"] for order in orders)
    print(f"{date} : ₹{daily_total:.2f}")

# Part 2 - Best selling day
best_day = max(sales_log, key=lambda date: sum(order["total"] for order in sales_log[date]))
best_revenue = sum(order["total"] for order in sales_log[best_day])
print(f"\nBest Day: {best_day} — ₹{best_revenue:.2f}")

# Part 3 - Most ordered item
item_count = {}
for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1

most_ordered = max(item_count, key=lambda x: item_count[x])
print(f"\nMost Ordered Item: {most_ordered} ({item_count[most_ordered]} orders)")

# Part 4 - Add new day and reprint
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

print("\n--- Updated Revenue Per Day ---")
for date, orders in sales_log.items():
    daily_total = sum(order["total"] for order in orders)
    print(f"{date} : ₹{daily_total:.2f}")

best_day = max(sales_log, key=lambda date: sum(order["total"] for order in sales_log[date]))
best_revenue = sum(order["total"] for order in sales_log[best_day])
print(f"\nUpdated Best Day: {best_day} — ₹{best_revenue:.2f}")

# Part 5 - Numbered list of all orders using enumerate
print("\n--- All Orders ---")
counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"{counter}.  [{date}] Order #{order['order_id']}  — ₹{order['total']:.2f} — Items: {items_str}")
        counter += 1
```
